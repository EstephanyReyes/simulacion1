Para esta prueba se consideran las siguientes hipótesis:

$H_0$ :Los números de la muestra $Xi$ son independientes (aleatorios en su orden)

$H_a$ : Los números de la muestra $Xi$ no son indendientes

Y se determina mediante los siguientes pasos:

Paso 1: Determinar la secuencia S de 0 y 1 ó (+ , -) así:


*  Si $xi \leq x_{i-1}$ escribir 0 ó $-$
* Si $xi > x_{i-1}$ escribir 1 ó $+$

La secuencia S contiene $n-1$ números

$$S={--+-+++--++-...}$$

Paso 2: Contar el número de rachas observadas $C_o$, una racha está dada por el número de ceros (-) o unos (+) consecutivos que la forman

Paso 3: Calcular el valor esperado y la varianza del número de rachas así como el estadístico $z_0$ mediante:
$$\mu_{C_o} = \frac{2n - 1}{3}$$
$$\sigma_{C_o}^{2} = \frac{16n - 29}{90}$$
$$
\sigma_{C_o} = \sqrt{\frac{16n - 29}{90}}
$$
$$
Z_0 = \frac{C_o - \mu_{C_o}}{\sigma_{C_o}}
$$

Dado que $c_O$ sigue una distribución normal de madia $\mu_{C_o}$ y varianza $\sigma_{C_o}$ y $z_0$ una distribución normal (0,1). El intervalo de aceptación es:
$$
- z_{\alpha/2} \leq z_0 \leq z_{\alpha/2}$$ esto implica
$$
\left| z_0 \right| \leq z_{\alpha/2}
$$





```python
import numpy as np
import pandas as pd
from scipy.stats import norm
```

```python
#importar datos
url = "https://raw.githubusercontent.com/EstephanyReyes/simulacion1/main/datos.txt"
datos = np.loadtxt(url)
```

```python
datos = np.array(datos)
print(datos)
```

```python


def prueba_rachas(datos, alpha=0.05):

    n = len(datos)

    # Paso 1: generar secuencia de signos
    signos = []
    for i in range(1, n):
        if datos[i] > datos[i-1]:
            signos.append("+")
        else:
            signos.append("-")

    # Paso 2: contar número de rachas observadas (Co)
    Co = 1
    for i in range(1, len(signos)):
        if signos[i] != signos[i-1]:
            Co += 1

    # Paso 3: media y varianza
    mu = (2*n - 1) / 3
    varianza = (16*n - 29) / 90
    desviacion = np.sqrt(varianza)

    # Paso 4: estadístico
    Z0 = (Co - mu) / desviacion

    # Paso 5: valor crítico
    z_critico = norm.ppf(1 - alpha/2)

    print("Número de datos (n):", n)
    print("Número de rachas observadas (Co):", Co)
    print("Media (mu):", mu)
    print("Varianza:", varianza)
    print("Z0:", Z0)
    print("Valor crítico:", z_critico)

    if abs(Z0) <= z_critico:
        print("No se rechaza H0 (los datos son independientes)")
    else:
        print("Se rechaza H0 (los datos no son independientes)")
```

```python
url = "https://raw.githubusercontent.com/EstephanyReyes/simulacion1/main/datos.txt"
datos = np.loadtxt(url)

prueba_rachas(datos)
```
