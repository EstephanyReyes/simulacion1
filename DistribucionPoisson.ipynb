#Distribución de Poisson


#Interpretación
La distribución de Poisson sirve para modelar el número de veces que ocurre un evento aleatorio durante un intervalo fijo (que puede ser de tiempo, distancia, área o volumen), bajo la condición de que conocemos su ritmo promedio de ocurrencia, pero no el momento exacto en el que sucederá cada evento.

Interpretación:

* El parámetro $(\lambda)$: Es la tasa promedio de éxito. Representa cuántas veces esperamos que ocurra el evento en ese intervalo. Si se cambia el tamaño del intervalo (por ejemplo, de una hora a dos horas), $\lambda$ cambia proporcionalmente.

* La variable $(X)$: Representa el número real de eventos que terminan ocurriendo en un intervalo específico. Mientras que $\lambda$ puede ser un número con decimales (como 2.5 accidentes por día), $X$ siempre será un número entero no negativo (\(0, 1, 2, 3...\)).

La forma de la distribución cambia drásticamente según el valor general de $\lambda$:

* Para un $\lambda$ pequeño $\lambda < 5$ : La distribución es asimétrica hacia la derecha. Esto significa que el valor más probable es bajo, y hay una "cola" larga hacia valores altos. Esto se interpreta esto como "eventos raros": la mayoría de las veces ocurrirán pocos eventos (o ninguno), pero existe una probabilidad pequeña de que ocurra un pico aislado de muchos eventos.

* A medida que $\lambda$ crece $\lambda > 15$: La distribución se vuelve simétrica y acampanada. Visualmente se empieza a parecer a una campana de Gauss (Distribución Normal). Interpreta esto como un flujo de eventos más constante y masivo; el resultado real fluctuará de forma equilibrada tanto por encima como por debajo del promedio.



#Utilidad


> Teoría de Colas y Dimensionamiento de Servicios (Líneas de Espera): Las empresas la usan para saber cuántos recursos necesitan para no colapsar.

* Centros de atención telefónica (Call Centers): calculan la probabilidad de que entren llamadas simultáneas para saber cuántos operadores contratar.

* Cajeros automáticos y Servidores Web: Determina cuántas peticiones de usuarios pueden llegar en un segundo para evitar que el sistema se caiga.

> Gestión de Riesgos y Seguros: Modelar eventos que ocurren con poca frecuencia pero que tienen un gran impacto.

* Accidentes de tráfico: Número de choques en un cruce de avenidas específico por mes.

* Reclamaciones de seguros: Cuántas pólizas por fallas estructurales o desastres naturales se cobrarán en un año.

> Control de Calidad y Manufactura: Evaluar la presencia de defectos en productos que se producen de manera continua.

* Textiles o Cables: Número de imperfecciones o rupturas por cada 100 metros de tela o cable producido.

* Líneas de producción: Número de productos defectuosos que salen de una banda transportadora por cada lote de mil unidades.

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
from math import exp
```

```python

# Función Poisson
def poisson(lam,n):
    muestra = []
    for k in range(n):
        U = r.random()
        i = 0
        p = exp(-lam)
        F = p
        while U >= F:
            p = (lam*p)/(i+1)
            F = F + p
            i = i + 1
        muestra.append(i)
    return muestra
```

```python

M = poisson(4,10000000)
plt.hist(M,bins=20)
plt.title("Distribución de Poisson")
plt.xlabel("x")
plt.ylabel("Frecuencia")
plt.show()
```
